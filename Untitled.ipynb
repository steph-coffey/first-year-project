```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

import importlib
import molmass

from main import CSRE, errorbars, chisquared
from plotting import plotting, sigma_lines, plot_depth_changes
```

```python
# name must match what is on name_params.py
systems = ['trappist1b', 'hats6b']
```

```python
# dictionary to store all the calculated quantities
# so I can use them later in notebook w/o rerunning
main_dict = {}

for name in systems:
    print(f'========== Starting  {name} ==========') # break up Pandexos printed output
    wave, spec, errs = CSRE(name)
    print(f'========== Done with {name} ==========')
    
    # dictionary of dictionaries
    system_dict = {'wave':wave,'spec':spec,'errs':errs}
    main_dict[f'{name}'] = system_dict
```

### $\chi^2$ vs $N_{transits}$ plot

```python
numtrans = np.linspace(1,400,400)
N = len(errs[-1])
chisq_vals = [] ; numsig_vals = []
for numtran in numtrans:
    chisq, numsig = chisquared('hats6b', errs, numtran)[0:2]
    chisq_vals.append(chisq) ; numsig_vals.append(numsig)
```

```python
fig, axs = plt.subplots(1,2, figsize = (10,4))

axs[0].plot(numtrans,chisq_vals, '.')
axs[0].set_xlabel('Number of transits')
axs[0].set_ylabel(r'$\chi^2$')

axs[1].plot(numtrans,numsig_vals, '.')
axs[1].set_xlabel('Number of transits')
axs[1].set_ylabel(r'Number of $\sigma$')
axs[1].set_xlim(1, np.max(numtrans))
axs[1].set_ylim(0, np.nanmax(numsig_vals))

fig.suptitle('HATS-6 b')

sigma_lines(fig, axs[0], numtrans, chisq_vals, N, [1,2,3,4])
```
