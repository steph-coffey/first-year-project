## Chisq

```python
x, y, yerr = errs
x, y, yerr = np.array(x), np.array(y), np.array(yerr)
```

```python
mu = np.sum(y * yerr**-2) / np.sum(yerr**-2)
mu
```

```python
chisq = np.sum(((y - mu)/yerr)**2)
chisq
```

```python
N = len(yerr)
numsig = np.sqrt(chisq - N)
numsig
```

```python
ax.axhline(mu, ls = 'dashed', c = 'seagreen', alpha = 0.65, label = 'flux weighted mean')
fig.text(0.765, 0.7, f'$\chi^2 = {int(chisq)}$', fontsize = 12, bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
fig
```
