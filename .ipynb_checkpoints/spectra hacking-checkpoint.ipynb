## this notebook is for our "hack" of ExoTransmit

**We're producing transit spectra for 2 different systems: a planet (hot Jupiter) transiting a Sun-like star, and an "M-dwarf" transiting the same Sun-like star. We then multiply out the stellar radius in the transit spectra for each, so we can then divide the planet spectra by the M-dwarf spectra.**

Note: all the spectra with ExoTransmit are calculated on the exact same wavelength grid, so I use the same wvelength variable throughout.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Can be found in kipping/Exo_Transmit/Spectra
# as of 19 Jan 2024
spectra_path = '/Users/coffey/Downloads/kipping/Exo_Transmit/Spectra'
```

### M2.5V Star: Ross 905
**T-P profile**                : 3000 K  
**EOS**                        : 1X_solar_gas  
**Planet g (m/s^2)**           : 653   
**Planet R (m)**               : 2.9219e8

I tried turning on collision induced opacities and Rayleigh scattering, but it had a marginal effect only at the very beginning of the wavelength range.

```python
mdwarf_spectrum  = np.loadtxt(f'{spectra_path}/transmission_mdwarf.dat', skiprows = 2).T

mdwarf_spec  = mdwarf_spectrum[1][:2900]

plt.figure(figsize = (8,4))
plt.title('Ross 905 transit spectrum')
plt.xlabel('Wavelength (microns)')
plt.ylabel('Transit Depth (percent)')

plt.plot(wave, mdwarf_spec)

plt.show()
```

## now we can multiply out the stellar radius for both

First we take the square root of the transit depth $\delta = (R_p/R_s)^2$, then multiply by $R_s = 7.6875*10^8 \ \text{m}$.

```python
Rs = 7.6875e8
Rh = 9.4655e7 # HAT-P-55 b
RM = 2.9219e8 # M dwarf

hatp55b_spec_noRs = np.sqrt(hatp55b_spec / 100) * Rs
mdwarf_spec_noRs  = np.sqrt(mdwarf_spec / 100) * Rs
```

```python
fig, axs = plt.subplots(2, 1, figsize = (9,6), sharex = True)

# Spectra
axs[0].plot(wave, hatp55b_spec_noRs, label = 'HAT-P-55 b')
axs[1].plot(wave, mdwarf_spec_noRs)
# Plotting their radii to confirm
axs[0].axhline(y = Rh, color = 'black', linestyle = 'dotted')
axs[1].axhline(y = RM, color = 'black', linestyle = 'dotted')

# Labels
fig.suptitle('Rp and RM spectra')
fig.text(0.04, 0.5, 'Transit depth (percent)', va='center', rotation='vertical')
fig.text(0.65, 0.825, 'HAT-P-55 b', bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
fig.text(0.65, 0.415, 'M dwarf', bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
axs[1].set_xlabel('Wavelength (microns)')
```

## combining the two spectra

Have $Rp$ and $Rs$ for our new $\delta = (R_p/R_s)^2$. Need to multiply by 100 to convert to percentage.

```python
hacked_spec = (hatp55b_spec_noRs / mdwarf_spec_noRs)**2 * 100
```

```python
fig, axs = plt.subplots(3, 1, figsize = (9,9), sharex = True)
fig.subplots_adjust(hspace = 0.1)

# Spectra
axs[0].plot(wave, hacked_spec, color = 'green')
axs[1].plot(wave, hatp55b_spec)
axs[2].plot(wave, mdwarf_spec)

"""# Plotting transit depths to compare
axs[0].axhline(y = (Rh/RM)**2 * 100, color = 'black', linestyle = 'dotted')
axs[1].axhline(y = (Rh/Rs)**2 * 100, color = 'black', linestyle = 'dotted')
axs[2].axhline(y = (RM/Rs)**2 * 100, color = 'black', linestyle = 'dotted')"""

# Labels
fig.suptitle('HAT-P-55 b orbiting an M dwarf', y = 0.915)
fig.text(0.04, 0.5, 'Transit depth (percent)', va='center', rotation='vertical')
fig.text(0.65, 0.575, 'HAT-P-55 b', bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
fig.text(0.65, 0.3, 'M dwarf', bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
axs[2].set_xlabel('Wavelength (microns)')
```

## for the final step we can smooth it to JWST prism resolution

```python
def adap_smooth(wave, spec, R):
    """
    Smooths to a fixed spectral resolution according to:
    
    R = lamb / deltalamb
    
    Thus, the bin size changes for each wavelength.
    
    ================== Params ==================
    
    wave   : array or list of wavelengths
    spec   : array or list of transit depths
    R      : spectral resolution (JWST NIRSPEC: ~100 (prism), ~1000, or ~2700)
    
    ============================================
    
    """
    
    # Finding center of first bin
    # using equation lamb = 2R*np.min(lambs) / (2R-1)
    lamb0 = (2*R*wave[0]) / (2*R - 1)
    dellamb = lamb0 / R
    
    # binning wavelengths w/n +- dellamb/2 of lamb0
    bin_indices = np.where(np.logical_and(wave >= lamb0 - dellamb/2, wave < lamb0 + dellamb/2))
    
    # taking mean in bin
    mean0 = spec[bin_indices].mean()
    
    # creating list to store the smooth spectrum
    smooth_spec = [mean0]
    
    # repeating everything above for every bin
    for i in range(1,len(spec)):
        lamb = wave[i]
        dellamb = lamb / R
        bin_indices = np.where(np.logical_and(wave >= lamb - dellamb/2, wave < lamb + dellamb/2))
        mean = spec[bin_indices].mean()

        smooth_spec.append(mean)
    
    return smooth_spec
```

```python
smooth_hacked_spec = adap_smooth(wave, hacked_spec, R = 100)
smooth_hatp55b_spec = adap_smooth(wave, hatp55b_spec, R = 100)
smooth_mdwarf_spec = adap_smooth(wave, mdwarf_spec, R = 100)
```

```python
fig, axs = plt.subplots(3, 1, figsize = (9,9), sharex = True)
fig.subplots_adjust(hspace = 0.1)

# Spectra
axs[0].plot(wave, smooth_hacked_spec, color = 'green')
axs[1].plot(wave, smooth_hatp55b_spec)
axs[2].plot(wave, smooth_mdwarf_spec)

# Plotting transit depths to compare
axs[0].axhline(y = (Rh/RM)**2 * 100, color = 'black', linestyle = 'dotted')
axs[1].axhline(y = (Rh/Rs)**2 * 100, color = 'black', linestyle = 'dotted')
axs[2].axhline(y = (RM/Rs)**2 * 100, color = 'black', linestyle = 'dotted')

# Labels
fig.suptitle('HAT-P-55 b orbiting an M dwarf', y = 0.915)
fig.text(0.04, 0.5, 'Transit depth (percent)', va='center', rotation='vertical')
fig.text(0.65, 0.575, 'HAT-P-55 b', bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
fig.text(0.65, 0.3, 'M dwarf', bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
axs[2].set_xlabel('Wavelength (microns)')
```
