```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.constants as scp
import astropy.constants as asp

import csv
```

$$ \large
\begin{align}
\Delta\delta_{star} &\simeq \gamma\delta \hspace{0.5cm}; \hspace{0.5cm} \gamma \equiv -2N\frac{H_*}{R_*} \\
\Delta\delta(M_*) &\propto \delta M_*^{\kappa} \hspace{0.5cm}; \hspace{0.5cm} \kappa \equiv \frac{\alpha}{2} + \frac{\beta}{2} - 1 
\end{align}
$$

```python
Re = asp.R_earth.value  # earth radius [m]
Me = asp.M_earth.value  # earth mass [kg]
Rsun = asp.R_sun.value  # solar radius [m]
Msun = asp.M_sun.value  # solar mass [kg]


def gamma(Rs, Ms, Ts, N):
    """
    Calculates gamma term in scale heights equation for a star.
    Assumes the mean molecular mass as that of Hydrogen.
     
        Rs    : radius of star (solar radii)
        Ms    : mass of star (solar masses)
        Ts    : temperature of star (solar units)
        N     : number of scale heights H (of order unity)  
    """
    
    # work in SI units
    Rs *= Rsun ; Ms *= Msun ; mu = scp.m_p # mean molecular mass (proton)
    
    # H = kT/mu*g
    g = scp.G * Ms / Rs**2
    Hs = (scp.k * Ts) / (mu * g)
    
    gamma = -2 * N * (Hs / Rs)
    
    return gamma


def deldel(gamma, depth):
    """
    Equation for change in scale heights from stellar atmosphere.
    
        gamma : return value of function gamma()
        depth : observed transit depth (fraction)
    """
    
    deldel = gamma * depth
    
    return deldel


def open_csv(filename):
    """
    Opens CSV file and puts data into array.
    
        filename : path to file (string)
    """
    
    # opening the CSV file
    with open(filename, mode ='r') as file:
        data = []
        # a bit easier to work with appending to table
        for line in csv.reader(file):
            print(line)
            data.append(line)
            
    return data
```

```python
table = open_csv('kepler_stars.csv')
#table = open_csv('kepler_Mdwarfs.csv')
```

```python
# finding the first row with data in csv (manually changing index)
print("Column names")
print("------------")
print(table[25])
print("")
print("First planet-star object")
print("------------------------")
print(table[26])
```

```python
# Splitting data into separate arrays for easier plotting
Rss = np.array([]) ; Mss = np.array([]); Tss = np.array([])
depths = np.array([])
deldelta = np.array([])
gammas = np.array([])

for planet in table[26:]:
   
    # solar units
    Rs = float(planet[-2]) # values are type 'str' in csv
    Ms = float(planet[-1])
    Ts = float(planet[-3])
    Rss = np.append(Rss,Rs)
    Mss = np.append(Mss,Ms)
    Tss = np.append(Tss,Ts)
    
    # ppm
    depth = float(planet[5])
    depths = np.append(depths,depth)
    
    # calculates gamma factor
    gma = gamma(Rs, Ms, Ts, N = 5)
    gammas = np.append(gammas,gma)
    
    # calculates change in transit depth
    dd = deldel(gma, depth)
    deldelta = np.append(deldelta,dd)
```

#### Calculating average, median, and std values

```python
gamma_med, gamma_std = np.median(gammas), np.std(gammas)
#print(f'Using {len(gammas)} confirmed transiting exoplanets around M-dwarfs,')
#print(f'we get gamma = {np.round(gamma_med, 5)} +- {np.round(gamma_std, 5)}')
```

```python
# Transit depth
fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, sharey = True, figsize = (15,5))
fig.subplots_adjust(wspace=0.05)

ax1.plot(Rss, depths, 'o')
ax2.plot(Mss, depths, 'o')
ax1.set_xlabel(r'Stellar radius $[R_{\odot}]$', fontsize = 15)
ax2.set_xlabel(r'Stellar mass $[M_{\odot}]$', fontsize = 15)
ax1.set_ylabel(r'$\delta$ (ppm)', fontsize = 15)

# for plotting kepler_stars
ax1.set_xlim(0,10)
ax1.set_ylim(0,100000)
ax2.set_ylim(0,100000)

plt.suptitle('Transit depth')
plt.show()
```

```python
# Change in transit depth
fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, sharey = True, figsize = (15,5))
fig.subplots_adjust(wspace=0.05)

ax1.plot(Rss, deldelta, 'o')
ax2.plot(Mss, deldelta, 'o')
ax1.set_xlabel(r'Stellar radius $[R_{\odot}]$', fontsize = 15)
ax2.set_xlabel(r'Stellar mass $[M_{\odot}]$', fontsize = 15)
ax1.set_ylabel(r'$\Delta\delta$ (ppm)', fontsize = 15)

# for plotting kepler_stars
ax1.set_xlim(0,10)
ax1.set_ylim(0,200)
ax2.set_ylim(0,200)

plt.suptitle('Change in transit depth')
plt.show()
```

```python
# Fractional change in transit depth
fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2, sharey = True, figsize = (15,5))
fig.subplots_adjust(wspace=0.05)

# this color is called 'Worcestershire Sauce'!
ax1.plot(Rss, gammas, 'o', color = 'mediumslateblue')
ax1.axhline(gamma_med, color = 'black', label = 'median')
ax1.axhline(gamma_med + gamma_std, linestyle = 'dashed', color = 'black', label = '1-sigma')
ax1.axhline(gamma_med - gamma_std, linestyle = 'dashed', color = 'black')

ax2.plot(Mss, gammas, 'o', color = 'mediumslateblue')
ax2.axhline(gamma_med, color = 'black')
ax2.axhline(gamma_med + gamma_std, linestyle = 'dashed', color = 'black')
ax2.axhline(gamma_med - gamma_std, linestyle = 'dashed', color = 'black')

# for plotting kepler_stars
ax1.set_xlim(0,20)
ax1.set_ylim(-0.02, 0)
ax2.set_ylim(-0.02, 0)

ax1.set_xlabel(r'Stellar radius $[R_{\odot}]$', fontsize = 15)
ax2.set_xlabel(r'Stellar mass $[M_{\odot}]$', fontsize = 15)
ax1.set_ylabel(r'$\Delta\delta / \delta$', fontsize = 15)
gamma_string = r'$\gamma$ = ' + f'{gamma_med:.4f} $\pm$ {gamma_std:.4f}'
ax1.text(10,-0.0035,gamma_string,)
ax1.legend()

#plt.suptitle('Fractional change in transit depth ')
plt.show()
```

### Error scaling

***HAT-p-18b to Trappist 1b***

```python
class planet:
    """
    Assigns planet object attributes to be used for scaling the error. 
    Helpful for being able to quickly scale between any two planets.
    
        name      : planet name (string)
        app_mag   : star's apparent magnitude
        trans_dur : transit duration [mins]
        sigma     : 1-sigma error
    """
    
    def __init__(self, name, app_mag, trans_dur, sigma):
        self.name      = name
        self.app_mag   = app_mag
        self.trans_dur = trans_dur
        self.sigma     = sigma
        
def scale(reference, planet):
    """
    Scales the error bars from one object to another,
    adjusting for brightness and transit duration.
    Must be from same telescope!
    
        reference : scaling from
        planet    : scaling to
    """
    
    m1 = reference.app_mag
    m2 = planet.app_mag
    B_ratio = 10**(-0.4*(m2 - m1))
    print('B_ratio:', B_ratio)
    
    t1 = reference.trans_dur
    t2 = planet.trans_dur
    t_ratio = t2 / t1
    print('t_ratio:', t_ratio)
    
    # scaling error from reference to planet
    sigma_planet = reference.sigma / np.sqrt(B_ratio) / np.sqrt(t_ratio)
    
    return sigma_planet
```

```python
# Opening data
hatp18b = np.array(open_csv('hat-p-18b.csv'))
trappist1b = np.array(open_csv('trappist-1b.csv'))

# Seperating the data and upper/lower error bars
med_h18b   = hatp18b[0::3, 1].astype(np.float64) 
upper_h18b = hatp18b[1::3, 1].astype(np.float64)
lower_h18b = hatp18b[2::3, 1].astype(np.float64)

med_t1b    = trappist1b[0::3, 1].astype(np.float64) 
upper_t1b  = trappist1b[1::3, 1].astype(np.float64)
lower_t1b  = trappist1b[2::3, 1].astype(np.float64)
```

```python
# averaging the error bars
errs_h18b = np.mean(np.array([upper_h18b - med_h18b, med_h18b - lower_h18b]), axis=0) # average of upper/lower
sigma_h18b = np.mean(errs_h18b) # average over wavelengths

errs_t1b = np.mean(np.array([upper_t1b - med_t1b, med_t1b - lower_t1b]), axis=0)
sigma_t1b = 1e6 * np.mean(errs_t1b) # converting to ppm

print('HAT-P-18b error:', sigma_h18b, 'ppm')
print('TRAPPIST-1b error:', sigma_t1b, 'ppm')
```

```python
# Defining planets w/ apparent magnitude and transit duration
h18b = planet('hatp18b', app_mag = 10.822, trans_dur = 162.864, sigma = sigma_h18b)
t1b = planet('trappist1b', app_mag = 15.6451, trans_dur = 36.06, sigma = sigma_t1b)
```

```python
t1b_estimate = scale(h18b, t1b)
print('Estimate of TRAPPIST-1b errors:', f'{t1b_estimate:.0f}', 'ppm')
print('Actual TRAPPIST-1b errors:', f'{t1b.sigma:.0f}', 'ppm')
```

```python
h18b_estimate = scale(t1b, h18b)
print('Estimate of HAT-p-18b errors:', f'{h18b_estimate:.0f}', 'ppm')
print('Actual HAT-p-18b errors:', f'{h18b.sigma:.0f}', 'ppm')
```
