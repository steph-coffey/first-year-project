### Chisq plot

```python
numtrans = np.linspace(1,1000,1000)
N = len(errs[-1])
chisq_vals = [] ; numsig_vals = []
for numtran in numtrans:
    chisq, numsig = chisquared('hats6b', errs, numtran)[0:2]
    chisq_vals.append(chisq) ; numsig_vals.append(numsig)
```

```python
def sigma_lines(fig, ax, x, y, N, numsig):
    """
    Handles the plotting of the lines for 
    1sigma, 2sigma, 3sigma, etc. on the chisq plot.
    
    """
    
    # plotting the line where chisq = N
    hline_value = N + numsig**2
    
    # min and max for lines
    hline_max = x[np.where(np.array(y)>=hline_value)[0][0]]
    vline_max = y[np.where(np.array(y)>=hline_value)[0][0]]
    
    ax.hlines(hline_value, 0, hline_max, ls = 'dotted', color = 'black')
    ax.vlines(hline_max, 0, vline_max, ls = 'dotted', color = 'black')
    
    ax.text(hline_max-4, hline_value - 15, f'${numsig} \sigma$', rotation=90, verticalalignment='center')
    
    return fig, ax
```

```python
fig, axs = plt.subplots(1,2, figsize = (10,4))

axs[0].plot(numtrans,chisq_vals)
axs[0].set_xlabel('Number of transits')
axs[0].set_ylabel(r'$\chi^2$')
axs[0].set_xlim(125,225)
axs[0].set_ylim(30, 80)
#axs[0].set_xlim(1, np.max(numtrans))
#axs[0].set_ylim(0, np.max(chisq_vals))

axs[1].plot(numtrans,numsig_vals)
axs[1].set_xlabel('Number of transits')
axs[1].set_ylabel(r'Number of $\sigma$')
axs[1].set_xlim(1, np.max(numtrans))
axs[1].set_ylim(0, np.nanmax(numsig_vals))

fig.suptitle('HATS-6 b')

for n in range(1,5):
    sigma_lines(fig, axs[0], numtrans, chisq_vals, N, n)
```
