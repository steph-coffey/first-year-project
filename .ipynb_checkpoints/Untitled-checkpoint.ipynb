```python
x,y,yerr = errorbars(wave, spec)
```

```python
yerr = yerr / np.sqrt(1e5)
```

```python
plt.plot(wave,spec)
plt.errorbar(x,y,yerr, None, '.')
plt.ylim(7358,7368)
```

```python
indices = []
for i in range(len(x)):
    index = (np.abs(wave - x[i])).argmin()
    indices.append(index)
```

```python
spec = np.array(spec)
```

```python
plt.plot(wave[indices],spec[indices], 'o')
plt.errorbar(x,y,yerr, None, '.')
plt.ylim(7358,7368)
```

## Chisq

```python
mu = np.sum(spec[indices] * yerr**-2) / np.sum(yerr**-2)
mu
```

```python
chisq = np.sum(((spec[indices] - mu)/yerr)**2)
chisq
```

```python
N = len(yerr)
numsig = np.sqrt(chisq - N)
numsig
```

```python
ax.axhline(mu, ls = 'dashed', c = 'seagreen', alpha = 0.65, label = 'flux weighted mean')
fig.text(0.765, 0.7, f'$\chi^2 = {int(chisq)}$', fontsize = 12, bbox=dict(boxstyle='round',ec=(1., 0.5, 0.5),fc=(1., 0.8, 0.8),))
fig
```
